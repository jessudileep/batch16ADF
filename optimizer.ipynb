```python
from pyspark.sql.functions import monotonically_increasing_id

# Define df as a sample DataFrame
df = spark.createDataFrame([(1, "Alice"), (2, "Bob")], ["number", "name"])

# Create a huge DataFrame by duplicating the existing df multiple times
huge_df = df.withColumn("id", monotonically_increasing_id())
for _ in range(500):
    huge_df = huge_df.union(df.withColumn("id", monotonically_increasing_id()))

# Cache the huge DataFrame
huge_df.cache()

# Materialize the cache
```

```python
display(huge_df.count())
```

```python
huge_df.display()
```

```python
simple_df = spark.createDataFrame([(1, "Alice"), (2, "Bob"), (3, "Carol")], ["number", "name"])
display(simple_df)

from pyspark.sql.functions import col

# Example: partitionBy when writing to disk
simple_df.write.mode("overwrite").partitionBy("number").parquet("/tmp/partitioned_output")

# Example: bucketing when writing to disk (requires saveAsTable)
simple_df.write.mode("overwrite").sortBy("name").saveAsTable("bucketed_table")

# Example: repartition to increase partitions
repartitioned_df = simple_df.repartition(10, col("number"))
display(repartitioned_df)

# Example: coalesce to reduce partitions
coalesced_df = simple_df.coalesce(2)
display(coalesced_df)
```
